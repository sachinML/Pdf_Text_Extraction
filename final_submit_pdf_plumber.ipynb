```python
import json
import pdfplumber

# Open the PDF file
with pdfplumber.open('D:\\Aidetic\\Mehul\\pdfjson\\Olive_Tree_Apartments_2550_Ogden_St__Denver__CO.pdf') as pdf_file:

    d = []
    # Loop over all pages of the PDF
    for page_num in range(len(pdf_file.pages)): 
        try:
            # Extract the text from the page
            page = pdf_file.pages[page_num]
            text = page.extract_text()

            # Extract the tables from the page
            tables = page.extract_tables()

            # Loop over all tables on the page
            for table_num, table in enumerate(tables):
                # Write the table data to the output file
                print(f"Table {table_num+1} on page {page_num+1}:")
                for row in table:
                    print(row)

                y = {'page no.': page_num+1,
                     'table no.': table_num+1,
                     'content': table}
                d.append(y)

            # Write the text to the output file
            print(f"Text on page {page_num+1}:")
            print(text)
            y = {'page no.': page_num+1,
                 'content': text}
            d.append(y)

        except Exception as e:
            # If an exception occurs, write the error message to the output file
            print("Error extracting data from page:", e)

        print("="*50)

# Write the data to a JSON file
with open("D:\\Aidetic\\Mehul\\pdfjson\\output\\Olive_Tree.json", "w", encoding="utf8") as output_file:
    json.dump(d, output_file, indent=4)
```
